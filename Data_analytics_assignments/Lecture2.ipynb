**Thoracic Surgery Data**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from scipy.io import arff
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
data_path = '/content/drive/MyDrive/Dateset/ThoraricSurgery.arff'
data, meta = arff.loadarff(data_path)
df = pd.DataFrame(data)
df_decoded = df.applymap(lambda x: x.decode('utf-8') if isinstance(x, bytes) else x)
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='DGN', data=df_decoded, palette='viridis')
plt.title('Distribution of Diagnoses')
plt.xlabel('Diagnosis')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df_decoded['AGE'], bins=20, kde=True, color='blue')
plt.title('Age Distribution of Patients')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='PRE4', y='PRE5', hue='Risk1Yr', data=df_decoded, palette='coolwarm')
plt.title('Scatter Plot of FEV1 vs Performance Status by Risk1Yr')
plt.xlabel('FEV1')
plt.ylabel('Performance Status')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Risk1Yr', data=df_decoded, palette='coolwarm')
plt.title('Risk Factor Distribution 1-Year Post Surgery')
plt.xlabel('Risk1Yr')
plt.ylabel('Count')
plt.show()
```
